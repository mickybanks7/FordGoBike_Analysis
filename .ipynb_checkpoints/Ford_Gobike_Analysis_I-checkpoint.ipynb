## Univariate Exploration

> First of, lets see the Distribution of time taken for rides (duration_sec) 


```python
# A histogram plot of duration_sec to get the frequency distribution
binsize = 500
bins = np.arange(0, df['duration_sec'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'duration_sec', bins = bins)
plt.title('Distribution of Duration in Seconds')
plt.xlabel('Duration (secs)')
plt.ylabel('Number of Trips')
plt.axis([-500, 12000, 0, 100000])
plt.show()
```

```python
# Distribution is skewd to the right. So let's put it on a log scale instead to get a proper view
log_binsize = 0.025
bins = 10 ** np.arange(2.4, np.log10(df['duration_sec'].max())+log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'duration_sec', bins = bins)
plt.xscale('log')
plt.xticks([500, 1e3, 2e3, 5e3, 1e4, 2e4], [500, '1k', '2k', '5k', '10k', '20k'])
plt.title('Log Distribution of Duration in Seconds')
plt.xlabel('Duration (Secs)')
plt.ylabel('Number of Trips')
plt.show()
```



> The first visualization had shown duration_sec to be long tailed and not well distributed. Using log scale, when plotted, duration looks modal and skewed to the right. You can see that higher peaks of trip appeared between 500 - 1000k before a steady decline began to occur.


> The birth year of members were given but the age wasn't. Up next, we will check the distribution of the member_birth_year.

```python
# plot a histogram of member_birth_year its distribution on number of trips

plt.figure(figsize=[8, 5])
plt.hist(data = df.dropna(), x = 'member_birth_year')
plt.title('Distribution of Birth Year')
plt.xlabel('Birth Year')
plt.ylabel('Number of Trips')
plt.show()
```

> The above graph is skewed to the left and shows that according to the member_birth_year, years within the new mellinum (2000) had more trips. We will do a feature engineering to get their age and also plot that distribution to see if the above graph and analysis is correct.

```python
# Age column was created, a histogram will be plotted to see distribution.
df['age'] = 2022 - df['member_birth_year']


binsize = 2
bins = np.arange(0, df['age'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df.dropna(), x = 'age')
plt.title('Distribution of Member Age')
plt.xlabel('Age (years)')
plt.ylabel('Number of Trips')
plt.show()
```

> If you recall, our member_birth_year distribution was skewed to the left (meaning persons from year 1980 down to recent years (2000) had more rides. The above plotted age distribution has confirmed that members below 50 had more rides and also, that riders between age 20 - 40 had the highest rides.


> Up next, I will feature engineer the start_time and end_time to get the start_day and end_day. A countplot with seaborn will be used to see the days with highest number of trips.

> From the graphs shown, it is evident that the correlation relationship between some features to another is below 50% (positive correlation) while some are more than 50% (negetively correlated) and less than a few are more than 50% (postively correlated)


> Lets pick some of these correlated features and plot a scatter plot of just the two features for more graphical view

```python
def scatterplot(data, xvariable, yvariable):
    '''
    This function is to be used to
    create a scatterplot that shows
    correlation between two features in
    our dataset using 400 samples for clear 
    and concise visuals
    '''
    #plt.figure is used to set the dimension of our plots
    plt.figure(figsize=[15,6])
    #lets plot the regplot using seaborn
    sb.regplot(data=df.sample(400), x=xvariable, y=yvariable);
    #plt.title gives the graph a title
    plt.title(f'correlation between {xvariable} and {yvariable}'.title(), fontsize=14, weight = "bold")
    #plt.xlable gives the graph a label on the x axis
    plt.xlabel(xvariable.title(), fontsize=10, weight="bold")
    #plt.yvariable gives the graph a label on the y axis
    plt.ylabel(yvariable.title(),fontsize=10, weight="bold")
```

```python
# lets see a visual of the correlation between start & end stations id
scatterplot(df, 'start_station_id', 'end_station_id')
```

> from the above graph, you can see that start_station_id has a positive correlation with end_station_id

> Up next, we plot a boxplot to see the distribution of few numeric columns with some categorical columns

```python
# using a sample size of 1000, lets create a boxplot of earlier grouped categorical columns against duration_sec and age
df_samp = df.sample(n=1000, replace = False)


def boxgrid(x, y, **kwargs):
    """ function for creating box plots with seaborn's PairGrid
    on a sample of 1000. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x=x, y=y, color=default_color)

plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = df_samp, y_vars = ['duration_sec', 'age'], x_vars = categoric_vars,
                height = 3, aspect = 1.8)
g.map(boxgrid)
plt.show()
```

```python
# plot function to plot not visible boxplots
base_color = sb.color_palette()[0]
def boxgraph(data, xvariable, yvariable, color):
    '''function to create a boxplot of duration_sec against other
    features for better understanding'''
    plt.figure(figsize = [10, 5])
    sb.boxplot(data = data, x = xvariable, y = yvariable, color = color)
    plt.ylim([-10, 5000])
    plt.xlabel(xvariable.title())
    plt.ylabel(yvariable.title())
    plt.title(f'Box plot of {xvariable} and {yvariable}'.title())
    plt.show()
```

```python
# boxplot of user_type against duration duration
boxgraph(df, 'user_type', 'duration_sec', base_color)
```

```python
# boxplot of member_gender against duration duration
boxgraph(df, 'member_gender', 'duration_sec', base_color)
```

## Multivariate Exploration

> Since we are more particular about trends that will cause a member to share bike, we will limit our plots to those who actually shared bike.

```python
# # lets convert the values in our target column from object to intger (! means a bike was shared and 0 means bike was not shared)
# df['bike_share_for_all_trip'] = df['bike_share_for_all_trip'].map({'Yes': 1, 'No': 0})
```

```python
# lets create a sub dataset from our dataset (seperate those who shared from those who did not)
shared_bike = df[df['bike_share_for_all_trip'] == 'Yes']
shared_not = df[df['bike_share_for_all_trip'] == 'No']
```

```python
#lets also create a numeric column of our target variable, so that we can plot easily
df['numeric_all_trip'] = (df['bike_share_for_all_trip'] == 'Yes').astype(int)
```

```python
# lets see the distribution of users that actually shared their bike using only the sub dataset for those who shared bike
user_type = sb.FacetGrid(data = shared_bike, col = 'user_type', col_wrap = 3, height = 4,
                    xlim = [10, 150], ylim = [-400, 9000])
user_type.map(plt.scatter, 'age', 'duration_sec', alpha = 0.2)
user_type.set_xlabels('Age (years)')
user_type.set_ylabels('Duration (sec)')
plt.show()
```

> Users that subscribe are the only ones to actually share a bike and most sharing occured below 60 years of age while members below 30 covered the most duration_sec in total and members around 50 having a high duration per trip. The oldest age that shared bike in the user category was around 125 there about.

```python
# lets see the distribution of duration and age on user_types that did not share a bike
user_type = sb.FacetGrid(data = shared_not, hue = 'user_type', height = 4,
                    aspect = 1.5, palette = 'viridis_r')
user_type.map(sb.regplot, 'age', 'duration_sec', x_jitter = 0.02, fit_reg = False)
user_type.add_legend()
user_type.set_xlabels('Age (years)')
user_type.set_ylabels('Duration (sec)')
plt.title('Distribution of Age and Duration_sec of all User_type that did not share their bike')
plt.show()
```

> It seems subcribed members also has the higher percentage of not sharing bike with most duration higher than that of customers type based on the above distribution. The customer type though has the oldest rider among all riders (age can be seen to be above 140)

```python
# lets see the distribution of users on all dataset using the numeric value of all trips as a marker
plt.scatter(data = df, x = 'user_type', y= 'duration_sec', c = 'numeric_all_trip', cmap = 'viridis_r')
plt.colorbar(label = 'bike_all_trip')
plt.xlabel('users type')
plt.ylabel('Duration (sec)')
plt.title('Distribution of User_type and Duration_sec on all Bike_trips')
plt.show()
```

> The above scatter plot with hue of all bike rides on the side, clearly show that all customer types are coloured yellow (0 == 'No') did not share a bike during their ride whereas if you look at the subscriber line, you can see that it's coloured both yellow and dark blue (where yellow means 0 and 'No' while dark blue means 1 and 'Yes") but though, in small quantity as aginst the 0 ('No') which means fewer riders in subcribers category actually shared a bike.

```python
# lets see the distribution of members_gender on those that actually shared a bike
rider_gender = sb.FacetGrid(data = shared_bike, col = 'member_gender', col_wrap = 3, height = 4,
                    xlim = [10, 90], ylim = [-400, 9000])
rider_gender.map(plt.scatter, 'age', 'duration_sec', alpha = 0.2)
rider_gender.set_xlabels('Age (years)')
rider_gender.set_ylabels('Duration (sec)')
plt.show()
```

> Earlier, we had seen female and other gender type having higher count in duration but then, in terms of those that actually shared a bike, male gender has the highest distribution and also, shows higher duration on trips.

```python
# lets see the distribution of start_day on those that did share a bike
rider_start_day = sb.FacetGrid(data = shared_bike, hue = 'member_gender', height = 4,
                    aspect = 1.5, palette = 'viridis_r')
rider_start_day.map(sb.regplot, 'age', 'duration_sec', x_jitter = 0.0002, fit_reg = False)
rider_start_day.add_legend()
rider_start_day.set_xlabels('Age (years)')
rider_start_day.set_ylabels('Duration (sec)')
plt.title('Distribution of Age and Duration_sec of Gender that shared a bike')
plt.show()
```

> As against what we saw during the bivarate visual analysis, it is clear that the male gender covered more durations on all trips based on its distribution but then, it is also clear that the older rider amongst all riders is a female.

> As against what we saw during the bivarate visual analysis, it is clear that the male gender covered more durations on all trips. 

```python
plt.figure(figsize = ([20,8]))
rider_start_day = sb.FacetGrid(data = shared_not, hue = 'end_day', height = 4,
                    aspect = 1.5, hue_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 
                                                                'Saturday', 'Sunday'])
rider_start_day.map(sb.regplot, 'age', 'duration_sec', x_jitter = 0.3, fit_reg = False)
rider_start_day.add_legend()
rider_start_day.set_xlabels('Age (years)')
rider_start_day.set_ylabels('Duration (sec)')
plt.title('Distribution of Age and Duration_sec on Weekdays of all riders that did not share a bike')
plt.show()
```

> A key point to note here is that the oldest rider on riders that did not share a bike rode on *Monday*

```python
# lets see the distribution of duration and age against a hue of all trips (either share or not)

plt.scatter(data = df, x = 'age', y= 'duration_sec', c = 'numeric_all_trip', cmap = 'viridis_r')
plt.colorbar(label = 'bike_all_trip')
plt.xlabel('Age (years)')
plt.ylabel('Duration (sec)')
plt.title('Distribution of Age and Duration_sec on all Bike_rides')
plt.show()
```
