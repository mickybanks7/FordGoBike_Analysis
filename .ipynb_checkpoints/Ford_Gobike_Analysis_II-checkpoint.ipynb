## Investigation Overview


> In this investigation, I wanted to see features that may prompt a rider to share a bike during a ride trip. The main features of focus was *'duration_sec'*, *'user_type'*, *'member_gender'*, *'member_birth_year'*. Three columns after been created by feature engineering existing columns were added to the list of main features (*'age'*, *'start_day'*, *'end_day'*)  


## Dataset Overview

> The data consisted of individual rides made in a bike-sharing system covering the greater San Francisco Bay area and it is centered on 2019 only. This document explores a dataset containing about 183,000 bike records and 16 features that clearly show details of how bikes are shared amongst riders.

> The dataset after all analysis and feature engineering had 20 columns while maintaining same length of rows.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
# The pandas to_datetime function will be used to convert the columns with wrong data types to its supposed data type
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
```

```python
# code that creates our supposed columns and also, get the day from each respective datetime column
df['start_day'] = df['start_time'].dt.day_name()
df['end_day'] = df['end_time'].dt.day_name()
```

## Distribution of Trip Duration (secs)

Trip duration_sec was long tailed and not well distributed. Using log scale, when plotted, duration looks modal and skewed to the right. You can see that higher peaks of trip appeared between 500 - 1000k before a steady decline began to occur.

```python
log_binsize = 0.025
bins = 10 ** np.arange(2.4, np.log10(df['duration_sec'].max())+log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'duration_sec', bins = bins)
plt.xscale('log')
plt.xticks([500, 1e3, 2e3, 5e3, 1e4, 2e4], [500, '1k', '2k', '5k', '10k', '20k'])
plt.title('Log Distribution of Duration in Seconds')
plt.xlabel('Duration (Secs)')
plt.ylabel('Number of Trips')
plt.show()
```

## Distribution of Member_birth_year

The histogram distribution was skewed to the left and shows that according to the year value of the memebers given, memebers years within the new mellinum (2000) had more trips. I did a feature engineering to get their age, the age distribution has confirmed that riders below 60 had more rides and also, that members between 20 - 30 had the highest rides.

```python
df['age'] = 2022 - df['member_birth_year']


binsize = 2
bins = np.arange(0, df['age'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df.dropna(), x = 'age')
plt.title('Distribution of Member Age')
plt.xlabel('Age (years)')
plt.ylabel('Number of Trips')
plt.show()
```

## Distribution of Start_day and End_day

The histogram distribution of start_day had seen thursday having the highest frequency of rides and exact same distribution was 
seen at the distribution of end_day.

```python
# plot start_day to see ride starting days with highest count of trips
base_color = sb.color_palette()[0]

plt.figure(figsize=[12, 5])
sb.countplot(x='start_day', data=df, color=base_color, order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 
                                                                'Saturday', 'Sunday'])
plt.title('Distribution of Ride Starting Days')
plt.xlabel('Weekday')
plt.ylabel('Number of Trips')
plt.show()
```

```python
# plot end_day to see ride ending days with highest count of trips
base_color = sb.color_palette()[0]


plt.figure(figsize=[12, 5])
sb.countplot(x='end_day', data=df, color=base_color, order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 
                                                                'Saturday', 'Sunday'])
plt.title('Distribution of Ride Ending Days')
plt.xlabel('Weekday')
plt.ylabel('Number of Trips')
plt.show()
```

## Correlation Plot of Entire Numeric Columns to Determine Relationship

There is less than 50% correlation between start_station_id and end_station_id while the relationship between member_birth_year and age is highly negetively correlated (100%). There is no correlation between duration_sec and start_station_id nor end_station_id. The relationship between start_station_latitude and end_station_latitude was about 99% (positive correlation. The correlation between start_station_latitude and end_station_longitude was more than 50% negetively correlated (-0.68) and same degree of correlation was seen between end_station_latitude and start_station_longtitude

```python
numeric_vars = ['duration_sec', 'start_station_id', 'start_station_latitude', 'start_station_longitude', 
                'end_station_id', 'end_station_latitude', 'end_station_longitude', 'bike_id', 'member_birth_year', 'age']

categoric_vars = ['user_type', 'member_gender', 'bike_share_for_all_trip']
```

```python
# correlation plot
plt.figure(figsize = [8, 5])
sb.heatmap(df[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

## Member_gender vs Bike_share_for_all_trip

The column of member_gender and bike_share_for_all_trip was plotted against each other to determine the relationship between them. The male gender had highest count when it came down to sharing a bike during the trip and equally topped the chart when options of not sharing a bike was looked at.

```python
# create 3 subplots using the full data.
plt.figure(figsize = [20, 16])

# # subplot 1: user_type vs bike_share_for_all_trip
# plt.subplot(3, 1, 1)
# sb.countplot(data = df, x = 'user_type', hue = 'bike_share_for_all_trip', palette = 'Blues')

# subplot 2: member_gender vs. bike_share_for_all_trip
ax = plt.subplot(3, 1, 2)
sb.countplot(data = df, x = 'member_gender', hue = 'bike_share_for_all_trip', palette = 'pink')
ax.legend(ncol = 2) # re-arrange legend to reduce overlapping
plt.title('Distribution of Member_gender vs Bike_share_for_all_trip')

# # subplot 3: member_gender vs. user_type, use different color palette
# ax = plt.subplot(3, 1, 3)
# sb.countplot(data = df, x = 'member_gender', hue = 'user_type', palette = 'Greens')
# ax.legend(loc = 1, ncol = 2) # re-arrange legend to remove overlapping

plt.show()
```

## Member_gender vs User_type

The both features we plotted against each other to determine the relationship of gender on user_type during the trip. Male gender subscribed most more than other gender. 

```python
# create 3 subplots using the full data.
plt.figure(figsize = [20, 16])

# # subplot 1: user_type vs bike_share_for_all_trip
# plt.subplot(3, 1, 1)
# sb.countplot(data = df, x = 'user_type', hue = 'bike_share_for_all_trip', palette = 'Blues')

# # # subplot 2: member_gender vs. bike_share_for_all_trip
# # ax = plt.subplot(3, 1, 2)
# # sb.countplot(data = df, x = 'member_gender', hue = 'bike_share_for_all_trip', palette = 'pink')
# # ax.legend(ncol = 2) # re-arrange legend to reduce overlapping

# subplot 3: member_gender vs. user_type, use different color palette
ax = plt.subplot(3, 1, 3)
sb.countplot(data = df, x = 'member_gender', hue = 'user_type', palette = 'Greens')
ax.legend(loc = 1, ncol = 2) # re-arrange legend to remove overlapping
plt.title('Distribution of Member_gender vs User_type')

plt.show()
```

## Duration_sec and Age on User_type

A FacetGrid mapped with scatterplot was plotted having duration_sec and age on user_type using option on *bike_share_for_all_trip == 'Yes'*, confirmed that riders between the age 20 - 30 covered higher durations and more distribution, only few riders in the subscriber user type shared their bikes. The oldest rider age that shared bike in the user category was around 125 years of age.

```python
shared_bike = df[df['bike_share_for_all_trip'] == 'Yes']
shared_not = df[df['bike_share_for_all_trip'] == 'No']
```

```python
#lets also create a numeric column of our target variable, so that we can plot easily
df['numeric_all_trip'] = (df['bike_share_for_all_trip'] == 'Yes').astype(int)
```

```python
# lets see the distribution of users that actually shared their bike
users = sb.FacetGrid(data = shared_bike, col = 'user_type', col_wrap = 3, height = 4,
                    xlim = [10, 150], ylim = [-400, 9000])
users.map(plt.scatter, 'age', 'duration_sec', alpha = 0.2)
users.set_xlabels('Age (years)')
users.set_ylabels('Duration (sec)')
plt.show()
```

```python
user_type = sb.FacetGrid(data = shared_bike, hue = 'user_type', height = 4,
                    aspect = 1.5, palette = 'viridis_r')
user_type.map(sb.regplot, 'age', 'duration_sec', x_jitter = 0.02, fit_reg = False)
user_type.add_legend()
user_type.set_xlabels('Age (years)')
user_type.set_ylabels('Duration (sec)')
plt.title('Distribution of Age and Duration_sec on Users that shared a bike ')
plt.show()
```

## Duration_sec and Age on Member_gender

A FacetGrid mapped with scatterplot was plotted having duration_sec and age on member_gender using option on *bike_share_for_all_trip == 'Yes'*. Male gender was seen to be well distributed with the male gender having more duration_sec than other gender types. During the exploration stage using countplot, we saw male gender also having high trips on all user types. A female rider covered the most duration_sec on a trip while the oldest rider bein a male.

```python
# lets see the distribution of members_gender on those that actually shared a bike
users = sb.FacetGrid(data = shared_bike, col = 'member_gender', col_wrap = 3, height = 4,
                    xlim = [10, 150], ylim = [-400, 9000])
users.map(plt.scatter, 'age', 'duration_sec', alpha = 0.2)
users.set_xlabels('Age (years)')
users.set_ylabels('Duration (sec)')
plt.show()
```

```python
# lets see the distribution of start_day on those that did share a bike
rider_start_day = sb.FacetGrid(data = shared_bike, hue = 'member_gender', height = 4,
                    aspect = 1.5, palette = 'viridis_r')
rider_start_day.map(sb.regplot, 'age', 'duration_sec', x_jitter = 0.0002, fit_reg = False)
rider_start_day.add_legend()
rider_start_day.set_xlabels('Age (years)')
rider_start_day.set_ylabels('Duration (sec)')
plt.title('Distribution of Age and Duration_sec of Gender that shared a bike during their ride')
plt.show()
```

## Duration_sec and Age on Bike_share_for_all_trip

The scatter plot clearly shows that most riders did not share a bike as the yellow colour indicates *0 == 'No' and also, that most older riders equally did not share a bike. It is also evident that less amount of riders shared a bike and thus, why the distribution of the yellow colour (0 == 'No') is more distributed.

```python
# lets see the distribution of duration and age against a hue of all trips (either share or not)

plt.scatter(data = df, x = 'age', y= 'duration_sec', c = 'numeric_all_trip', cmap = 'viridis_r')
plt.colorbar(label = 'bike_all_trip')
plt.xlabel('Age (years)')
plt.ylabel('Duration (sec)')
plt.title('Distribution of Age and Duration_sec on all Bike_rides')
plt.show()
```
